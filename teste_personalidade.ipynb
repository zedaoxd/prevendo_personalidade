deletando colunas que não vou usar, só quero as referente as perguntas

deletando as linhas com valor zerado

```python
df_personalidade = df_personalidade[(df_personalidade > 0.00).all(axis=1)]
```

```python
df_personalidade.shape
```

```python
from sklearn.cluster import KMeans
```

dividindo os registros nos 5 grupos

```python
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(df_personalidade)
```

```python
predicoes = k_fit.labels_
df_personalidade['Clusters'] = predicoes
```

```python
df_personalidade.columns = df_personalidade.columns.str.replace('Clusters','Personalidade')
df_personalidade.head()
```

```python
df_personalidade['Personalidade'].value_counts()
```

```python
df_personalidade.groupby('Personalidade').mean()
```

selecionando as colunas de cada tipo de pergunta

```python
lista_coluna = list(df_personalidade)
ext = lista_coluna[0:10]
est = lista_coluna[10:20]
agr = lista_coluna[20:30]
csn = lista_coluna[30:40]
opn = lista_coluna[40:50]
```

somando os valores de cada grupo

```python
df_soma = pd.DataFrame()
df_soma['extrovertido'] = df_personalidade[ext].sum(axis=1)/10
df_soma['neurotico'] = df_personalidade[est].sum(axis=1)/10
df_soma['agressivo'] = df_personalidade[agr].sum(axis=1)/10
df_soma['consciencioso'] = df_personalidade[csn].sum(axis=1)/10
df_soma['aberto'] = df_personalidade[opn].sum(axis=1)/10
df_soma['Personalidade'] = predicoes
```

```python
df_soma.groupby('Personalidade').mean()
```

```python
data_personalidade = df_soma.groupby('Personalidade').mean()
```

```python
plt.figure(figsize=(23, 2))
lista = ['extrovertido', 'neurotico', 'agressivo', 'consciencioso' ,'aberto']
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.bar(data_personalidade.columns, data_personalidade.iloc[:, i], color='green', alpha=0.2)
    plt.plot(data_personalidade.columns, data_personalidade.iloc[:, i], color='red')
    plt.title(str(lista[i]))
    plt.xticks(rotation=45)
    plt.ylim(0, 4);
```

```python
questao_da_vez = []
for i in questoes:
    obj_input = gr.inputs.Radio(['1', '2', '3', '4', '5'],label=i)
    questao_da_vez.append(obj_input)
```

```python
questao_da_vez
```

```python
def predict(*args):
    args = np.array(args).reshape(1, -1)
    previsao = k_fit.predict(args)
    if previsao[0] == 0:
        return 'extrovertido'
    elif previsao[0] == 1:
        return 'neurotico'
    elif previsao[0] == 2:
        return 'agressivo'
    elif previsao[0] == 3:
        return 'consciencioso'
    elif previsao[0] == 4:
        return 'aberto a oportunidades'

iface = gr.Interface(
    fn = predict,
    title = 'Teste de Personalidade',
    description = 'Detecta a personalidade de uma pessoa',
    inputs = questao_da_vez,
    outputs = 'text'
)
iface.launch(share=True)
```
